```python
class Ucus():
    havaYolu = "THY"
    def __init__(self, kod, kalkisYeri, varisNoktasi, sure, kapasite, yolcu):
        self.kod = kod
        self.kalkisYeri = kalkisYeri
        self.varisNoktasi = varisNoktasi
        self.sure = sure
        self.kapasite = kapasite
        self.yolcu = yolcu
    def __repr__(self):
        return f"{self.kod} sefer sayılı uçuş sisteme tanımlanmıştır."
    def anonsYap(self):
        return f"{self.kod} sefer sayılı {self.kalkisYeri}-{self.varisNoktasi} uçuşumuz {self.sure} dakika sürecektir!"   
    def koltukSayisiGuncelle(self):
        return self.kapasite - self.yolcu
    def biletSatis(self, biletAdedi=1):
        if self.yolcu + biletAdedi <= self.kapasite:
            self.yolcu += biletAdedi
            self.koltukSayisiGuncelle()
            return(f"{biletAdedi} adet bilet satılmıştır. Kalan koltuk sayısı {self.koltukSayisiGuncelle()}")
        else:
            print("Yetersiz kapasite")
    def biletIptal(self, iptalAdedi =1):
        if self.yolcu >= iptalAdedi:
            self.yolcu -= iptalAdedi
            self.koltukSayisiGuncelle()
            return(f"{iptalAdedi} adet bilet iptal edilmiştir. Kalan koltuk sayısı {self.koltukSayisiGuncelle()}")
        else:
            print("İşlem gerçekleştirilemedi. İptal edilecek kadar yolcu yoktur")
        
```

```python
ucus3 = Ucus("TK323","BOD","ANT",40,250,200)
```

```python
ucus3.koltukSayisiGuncelle()
```

```python
ucus3.biletSatis(5)
```

```python
ucus3.biletSatis(25)
```

```python
ucus3.biletSatis(30)
```

```python
ucus3.biletIptal(20)
```

```python
ucus3.biletIptal(300)
```

#### **Inheritance**

```python
class Seyahat():
    def __init__(self, kalkisYeri, varisYeri):
        self.kalkisYeri = kalkisYeri
        self.varisYeri = varisYeri
    def anons(self):
        return f"{self.kalkisYeri}-{self.varisYeri} seyahatine hoşgeldiniz"
class Otobus(Seyahat):
    def __init__(self, molaDuraklari, kalkisYeri, varisYeri):
        Seyahat.__init__(self, kalkisYeri, varisYeri)
        self.molaDuraklari = molaDuraklari
```

```python
seyahat1 = Seyahat("ANT","BOD")
```

```python
seyahat1.kalkisYeri
```

```python
seyahat1.anons()
```

```python
oto1 = Otobus(["FETH","ALAN"], "ANT","BOD")
```

```python
oto1.molaDuraklari
```

```python
oto1.anons()
```
